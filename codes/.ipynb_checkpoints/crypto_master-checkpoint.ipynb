```python
# The process of reading csv file and renaming is repetitive, thus define a function read_crypto_price(str) to do it
# Parameter 'str' means the ticker for the crypto and naming of the CSV file should be the same with the ticker
# This functions returns the dataframe of the CSV file with the close prices of the crypto

def read_crypto_price(str):
    path_str = "../Resources/"+str+".csv"
    path = Path(path_str)
    data = pd.read_csv(path, index_col = 'Date', parse_dates = True,infer_datetime_format = True,dayfirst=True)
    data.index = pd.to_datetime(data.index.date)
    rename_str = str + " Close"   
    data = data.rename(columns={'Close':rename_str})

    return data
```

# Crypto Selection
- According to market cap data from yahoo finance 
- Exclude cryptos closely linked to USD, e.g. USDT, USDC etc.
- Top 6 Cryptos which have the largest market cap

```python
ada_data = read_crypto_price('ADA')
btc_data = read_crypto_price('BTC')
eth_data = read_crypto_price('ETH')
sol_data = read_crypto_price('SOL')
xrp_data = read_crypto_price('XRP')
bnb_data = read_crypto_price('BNB')

btc_data.describe()
```

```python
# Calculate Daily Returns
combined_daily_return = combined_price.pct_change()

# Rename the column to 'XXX D_Return'
combined_daily_return = combined_daily_return.rename(
    columns=
    {
        'BTC Close':'BTC D_Return',
        'ETH Close':'ETH D_Return',
        'BNB Close':'BNB D_Return',
        'XRP Close':'XRP D_Return',
        'ADA Close':'ADA D_Return',
        'SOL Close':'SOL D_Return'
    }
)
combined_daily_return.dropna(inplace=True)
combined_daily_return
```

```python
# plot the daily return of the 6 cryptos
combined_daily_return.plot(figsize=(20,10),title='Daily Return of All Cryptos')
```

## Observation and Analysis

```python
# Calculate cumulative returns of the 6 cryptos
cumulative_return = (1 + combined_daily_return).cumprod()-1

cumulative_return
```

```python
# Calculate the daily standard deviations

combined_std = combined_daily_return.std()
combined_std
```

```python
combined_volatility = combined_std * np.sqrt(252)
```

```python
# Plot the Standard Deviation of the 6 Cryptos

combined_volatility.plot(kind = 'bar',title='Standard Deviation of the 6 Cryptos')
```

```python
sharpe_ratios = (combined_daily_return.mean() * 252) / combined_volatility
sharpe_ratios.plot(kind = 'bar',title='Sharpe Ratios of the 6 Cryptos')
```

# Comparison between _Dollar-Cost Averaging_ and _Lump Sum Investment_ in 5 years time range

```python
def dca_lump_compare(coin_data,ticker,monthly_inv_amount,months):

    # ****************************************************************************
    # A DCA Investment Return Calculation
    # ****************************************************************************
    
    # Within 5 years time period, get the close price once every month, calculate the amount by 100/close_price    
    # Get the sum amount and calculate the value by the final (close price * sum_amount)

    # Extract the past 5 years' data from coin_data, time range starts from '2017-09-01' and end '2022-08-01' totally 60 months
    coin_data_slice = coin_data['2017-09-01':'2022-08-01']
    coin_data_slice

    # Group by year and month and get the first price of each month using first() function
    coin_inv_price=coin_data_slice.groupby([coin_data_slice.index.year, coin_data_slice.index.month]).first()
    coin_inv_price
    # Group by year and month and get 

    # Calculate the total amount acquried for the $100 monthly investment

    initail_amount = monthly_inv_amount * months
    amount = 0
    total_amount = 0
    i = 0

    coin_inv_price.reset_index(inplace=True)
    coin_inv_price
    
    # Get row counts
    df = coin_inv_price.describe()
    count = int(df.iloc[0,0])

    print('***********************************************\n')
    print(ticker)
    print('\n***********************************************')  

    while i<count:
        price = coin_inv_price.iloc[i,2]
        amount = round(monthly_inv_amount/price,4)
        total_amount = round((total_amount + amount),4)
        i += 1

    dca_value = total_amount * price

    # investment over 5 years total at $100*60 = $6000
    dca_profit = dca_value - initail_amount


    # ****************************************************************************
    # A Lump Sum Investment Return Calculation
    # ****************************************************************************

    coin_daily_return = coin_data_slice.pct_change()
    coin_cumprod = (coin_daily_return+1).cumprod()-1
    coin_cumprod.dropna(inplace=True)   
    
    column_name = ticker + ' Close'
    cum_return_coin = coin_cumprod.loc['2022-08-01',column_name]
    lump_sum_profit = initail_amount * cum_return_coin

    print(f'DCA profit: {dca_profit}')
    print(f'Lump Sum profit: {lump_sum_profit}')
    
    compare_list = [dca_profit,lump_sum_profit]
    
    return compare_list

```

```python
compare_list_btc = ['BTC'] + dca_lump_compare(btc_data,'BTC',100,60)
compare_list_ada = ['ADA'] + dca_lump_compare(ada_data,'ADA',100,60)
compare_list_sol = ['SOL'] + dca_lump_compare(sol_data,'SOL',100,60)
compare_list_xrp = ['XRP'] + dca_lump_compare(xrp_data,'XRP',100,60)
compare_list_bnb = ['BNB'] + dca_lump_compare(bnb_data,'BNB',100,60)
compare_list_eth = ['ETH'] + dca_lump_compare(eth_data,'ETH',100,60)

compare_list = [compare_list_btc,compare_list_eth,compare_list_ada,compare_list_sol,compare_list_xrp,compare_list_bnb]  

compare_list
```

```python
df = pd.DataFrame(np.array(compare_list), columns = ['Ticker','DCA','Lump Sum']).set_index('Ticker')

#convert the  column data type into float
df = df.astype({"DCA": float,"Lump Sum":float})

df
```

```python
df.plot(kind = 'bar',title='DCA Vs. Lump Sum')
```

## Observation and Analysis
- Lump Sum > DCA: BTC, ADA, XRP, SOL, BNB
- DCA > Lump Sum: ETH

- We cannot get to the conclusion which investment strategy is absolutely better than the other one
- However, according to the 6 cryptos we selected, over 80% of the time, Lump Sum Investment wins and sometimes wins in a great deal
- The comparison can be more comprehensive if we compare the sharpe ratio and standard deviation of the two (Taking into consideration of the risk factors), but due to time limitation, these are not covered in this project.
